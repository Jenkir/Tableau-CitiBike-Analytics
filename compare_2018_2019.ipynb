```python
# Add dependencies

import pandas as pd
import requests
import numpy as np
import datetime 
import os
import glob
```

```python
path = "raw_data"
all_files = glob.glob(path + "/*.csv")
merged_df = pd.DataFrame()
timeYr = []

for file_ in all_files:
    year_df = pd.read_csv(file_,index_col=None, header=0)
    timeYr.append(year_df)
merged_df = pd.concat(timeYr)
#frame.to_csv("data/data_merged.csv", index=False)
merged_df.head()
```

```python
#see example below of how to select columns you want to keep in a df without changing original df(s) by using .copy()
#new_customer_data_df = customer_data_df[['id', 'first_name', 'last_name']].copy()
#new_customer_data_df.head()
```

```python
#Add "Minutes" column for trip duration
merged_df['Duration(Minutes)'] = merged_df['tripduration']/60
#For Visual purposes, rounded
merged_df['Duration(Minutes)'] = round(merged_df['Duration(Minutes)'])
merged_df['Duration(Minutes)'] = merged_df['Duration(Minutes)'].astype(int)

```

```python
merged_df.tail()
```

```python
#len gives number of rows
len(merged_df)
```

```python
#take sample of the data so that files are not too large 
sample_df = merged_df.sample(frac=0.25)
sample_df.tail()
```

```python
len(sample_df)
```

```python
# Delete extraneous column
#merged_df = merged_df.drop(['usertype'], axis=1)
```

```python
rename_df = sample_df.rename(columns={"tripduration":"Trip Duration", "starttime":"Start Time", "stoptime": "End Time",
                                                       "start station id":"Start Station ID", "start station name":"Start Station Name","startstationlatitude":"Start Station Latitude",
                                                      "start station longitude":"Start Station Longitude", "end station id":"End Station ID", "end station name":"End Station Name", 
                                                      "end station latitude":"End Station Latitude", "end station longitude":"End Station Longitude", "bikeid":"Bike ID", "birth year":"Birth Year",
                                                      "usertype": "User Type", "gender": "Gender", "Duration(Minutes)": "Duration (Minutes)"})
rename_df.tail()
```

```python
# Delete extraneous columns
rename_df = rename_df.drop(['Trip Duration'], axis=1)
rename_df.head()
```

```python
# Check for missing values
rename_df.count()
```

```python
# drop null rows
no_null_df = rename_df.dropna(how='any')
len(no_null_df)
```

```python
no_null_df['Year'] = pd.to_datetime(no_null_df['Start Time']).dt.year
#final_2018_2019_df['Month'] = pd.to_datetime(final_2018_2019_df['Start Time']).dt.month
#final_2018_2019_df['Day'] = pd.to_datetime(final_2018_2019_df['Start Time']).dt.day
#final_2018_2019_df['Hour'] = pd.to_datetime(final_2018_2019_df['Start Time']).dt.hour
#final_2018_2019_df['Time'] = pd.to_datetime(final_2018_2019_df['Start Time']).dt.time
no_null_df.tail()
```

```python
#Use pandas groupby function to find most popular trips
popular_df = no_null_df.groupby(['Start Station Name', 'End Station Name']).size().reset_index(name = 'Number of Trips')
popular_df.tail()
```

```python
final_df = pd.merge(no_null_df, popular_df, how='inner')
                    #['Number of Trips'], on=["Start Station Name", "End Station Name"], how='left')
final_df.head()
```

```python
# Export to csv
final_df.to_csv("clean_data/final_compared_2018_2019.csv")
```
